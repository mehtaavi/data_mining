```python
#starting cluster numbering with 1 instead of 0

kmeans_1 = x_kmeans
kmeans_1 = x_kmeans + 1
```

```python
#  new dataframe with the name cluster

cluster = pd.DataFrame(kmeans_1)
```

```python
# Adding cluster to the Dataset1

k_means_data['cluster'] = cluster
```

```python
#Mean of clusters

kmeans_mean_cluster = pd.DataFrame(round(k_means_data.groupby('cluster').mean(),1))
kmeans_mean_cluster
```

```python
def cluster_percentage(data,name):
    """function to calculate the percentage and size of cluster"""
    
    size = pd.Series(data[name].value_counts().sort_index())
    percent = pd.Series(round(data[name].value_counts() / data.shape[0] * 100,2)).sort_index()

    size_df = pd.concat([size, percent],axis=1)
    size_df.columns = ["Cluster_Size" , "Cluster_Percentage"]
    
    return(size_df)
```

```python
cluster_percentage(k_means_data, "cluster")
```
